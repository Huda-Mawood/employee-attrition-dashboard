### Exploratory Data Analysis (EDA)

```python
# Does Distance from Home affect attrition

fig = px.box(
    df,
    x="attrition",
    y="distance_from_home",
    color="attrition",
    title="Distance from Home vs Attrition"
)

fig.show()
```

```python
#Does Age affect employee attrition?

fig = px.box(
    df,
    x="attrition",
    y="age",
    color="attrition",
    title="Age vs Employee Attrition"
)

fig.show()
```

```python
# Does number of promotions affect employee attrition?
promotions_attrition = (
    df.groupby("number_of_promotions")["attrition"]
    .value_counts(normalize=True)
    .reset_index(name="ratio")
)

left_only = promotions_attrition[promotions_attrition["attrition"] == 1]

fig = px.line(
    left_only,
    x="number_of_promotions",
    y="ratio",
    markers=True,
    title="Dose-Response: Promotions vs Attrition Rate"
)

fig.show()
```

```python
# Does Gender affect attrition
gender_attrition = df.groupby("gender")["attrition"].value_counts(normalize=True).reset_index(name="ratio")

fig = px.bar(
    gender_attrition,
    x="gender",
    y="ratio",
    color="attrition",
    barmode="group",
    title="Gender vs Attrition Rate"
)

fig.show()
```

```python
# Does Monthly Income affect attrition
fig = px.violin(
    df,
    x="attrition",
    y="monthly_income",
    box=True,
    color="attrition",
    title="Monthly Income vs Attrition"
)

fig.show()
```

```python
# Does Monthly Income affect attrition
fig = px.box(
    df,
    x="attrition",
    y="monthly_income",
    color="attrition",
    title="Monthly Income vs Attrition"
)

fig.show()
```

```python
# Does Years at Company affect attrition
size_attrition = df.groupby("company_size")["attrition"].value_counts(normalize=True).reset_index(name="ratio")

fig = px.bar(
    size_attrition,
    x="company_size",
    y="ratio",
    color="attrition",
    barmode="group",
    title="Company Size vs Attrition"
)

fig.show()
```

```python
remote = (
    df.groupby("remote_work")["attrition"]
    .value_counts(normalize=True)
    .reset_index(name="ratio")
)

fig = px.bar(
    remote,
    x="remote_work",
    y="ratio",
    color="attrition",
    barmode="group",
    title="Remote Work vs Attrition Rate"
)

fig.show()
```

```python
# Who leaves the company the most

ct = pd.crosstab(df["job_role"], df["attrition"], normalize="index")

fig = px.imshow(
    ct,
    text_auto=True,
    title="Attrition Rate by Job Role"
)

fig.show()
```

```python
# Do Work-Life Balance and Remote Work affect attrition
wlb = df.groupby("work-life_balance")["attrition"].value_counts(normalize=True).reset_index(name="ratio")

fig = px.bar(
    wlb,
    x="work-life_balance",
    y="ratio",
    color="attrition",
    barmode="group",
    title="Work-Life Balance vs Attrition"
)

fig.show()
```

```python
poor_fair = df[df["work-life_balance"].isin(["Poor", "Fair"])]

pd.crosstab(
    poor_fair["overtime"],
    poor_fair["attrition"],
    normalize="index"
)
```

```python
poor_fair = df[df["work-life_balance"].isin(["Poor", "Fair"])]

satisfaction = (
    poor_fair.groupby("job_satisfaction")["attrition"]
    .value_counts(normalize=True)
    .reset_index(name="ratio")
)

fig = px.bar(
    satisfaction,
    x="job_satisfaction",
    y="ratio",
    color="attrition",
    barmode="group",
    title="Job Satisfaction vs Attrition (Poor/Fair Work-Life Balance)"
)

fig.show()
```

```python
recognition = (
    poor_fair.groupby("employee_recognition")["attrition"]
    .value_counts(normalize=True)
    .reset_index(name="ratio")
)

fig = px.bar(
    recognition,
    x="employee_recognition",
    y="ratio",
    color="attrition",
    barmode="group",
    title="Employee Recognition vs Attrition"
)

fig.show()
```

```python
leadership = (
    poor_fair.groupby("leadership_opportunities")["attrition"]
    .value_counts(normalize=True)
    .reset_index(name="ratio")
)

fig = px.bar(
    leadership,
    x="leadership_opportunities",
    y="ratio",
    color="attrition",
    barmode="group",
    title="Leadership Opportunities vs Attrition"
)

fig.show()
```

```python
recognition = (
    df.groupby("employee_recognition")["attrition"]
    .value_counts(normalize=True)
    .reset_index(name="ratio")
)

fig = px.bar(
    recognition,
    x="employee_recognition",
    y="ratio",
    color="attrition",
    barmode="group",
    title="Employee Recognition vs Attrition"
)

fig.show()
```

```python
df["attrition"] = df["attrition"].map({
    "Stayed": 0,
    "Left": 1
})
numeric_df = df.select_dtypes(include=["int64"])
corr = numeric_df.corr()

plt.figure(figsize=(12,8))
sns.heatmap(corr, annot=False, cmap="coolwarm", center=0)
plt.title("Correlation Heatmap")
plt.show()
```

```python
high_risk = df[
    (df["remote_work"] == "No") &
    (df["distance_from_home"] > df["distance_from_home"].median()) &
    (df["work-life_balance"].isin(["Poor", "Fair"])) &
    (df["number_of_promotions"] == 0)
]

risk = (
    high_risk["attrition"]
    .value_counts(normalize=True)
    .reset_index(name="ratio")
)

fig = px.pie(
    risk,
    names="attrition",
    values="ratio",
    title="High Risk Group Attrition Distribution"
)

fig.show()
```

```python
low_risk = df[
    (df["remote_work"] == "Yes") &
    (df["distance_from_home"] <= df["distance_from_home"].median()) &
    (df["work-life_balance"].isin(["Good", "Excellent"])) &
    (df["number_of_promotions"] > 0)
]

risk = (
    low_risk["attrition"]
    .value_counts(normalize=True)
    .reset_index(name="ratio")
)

fig = px.pie(
    risk,
    names="attrition",
    values="ratio",
    title="Low Risk Group Attrition Distribution"
)

fig.show()
```

### Visualization 

#### Overall Attrition Distribution

```python
# Who is leaving the company overall
attrition_counts = df["attrition"].value_counts().reset_index()
attrition_counts.columns = ["attrition", "count"]

fig = px.pie(
    attrition_counts,
    names="attrition",
    values="count",
    title="Overall Attrition Distribution"
)

fig.show()
```

#### Work-Life Balance vs Attrition

```python
# Does work-life balance affect attrition
wlb = df.groupby("work-life_balance")["attrition"] \
    .value_counts(normalize=True) \
    .reset_index(name="ratio")

fig = px.bar(
    wlb,
    x="work-life_balance",
    y="ratio",
    color="attrition",
    barmode="group",
    title="Work-Life Balance vs Attrition"
)

fig.show()
```

#### Remote Work Effect

```python
# Does remote work reduce attrition
remote = df.groupby("remote_work")["attrition"] \
    .value_counts(normalize=True) \
    .reset_index(name="ratio")

fig = px.bar(
    remote,
    x="remote_work",
    y="ratio",
    color="attrition",
    barmode="group",
    title="Remote Work vs Attrition"
)

fig.show()
```

#### Promotions vs Attrition (Dose Effect)

```python
# Do promotions reduce attrition
prom = df.groupby("number_of_promotions")["attrition"] \
    .value_counts(normalize=True) \
    .reset_index(name="ratio")

left_only = prom[prom["attrition"] == 1]

fig = px.line(
    left_only,
    x="number_of_promotions",
    y="ratio",
    markers=True,
    title="Promotions vs Attrition Rate"
)

fig.show()
```

#### High Risk vs Low Risk

```python
# What happens when multiple risk factors combine
high_risk = df[
    (df["remote_work"] == "No") &
    (df["distance_from_home"] > df["distance_from_home"].median()) &
    (df["work-life_balance"].isin(["Poor", "Fair"])) &
    (df["number_of_promotions"] == 0)
]

risk = high_risk["attrition"].value_counts(normalize=True).reset_index(name="ratio")

fig = px.pie(
    risk,
    names="attrition",
    values="ratio",
    title="High Risk Group Attrition"
)

fig.show()
```

```python
low_risk = df[
    (df["remote_work"] == "Yes") &
    (df["distance_from_home"] <= df["distance_from_home"].median()) &
    (df["work-life_balance"].isin(["Good", "Excellent"])) &
    (df["number_of_promotions"] > 0)
]

risk2 = low_risk["attrition"].value_counts(normalize=True).reset_index(name="ratio")

fig = px.pie(
    risk2,
    names="attrition",
    values="ratio",
    title="Low Risk Group Attrition"
)

fig.show()
```
